```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head() #Display column headers for later reference
```


## Player Count

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Another simple dataframe for data and metadata
item_dat = purchase_data[["Purchase ID","Item Name","Price"]]
itm_res = pd.DataFrame({"Number of Unique Items":[len(item_dat["Item Name"].unique())],
                       "Average Price":[round(item_dat["Price"].mean(),2)],
                       "Number of Purchases":[item_dat["Purchase ID"].count()],
                       "Total Revenue":[item_dat["Price"].sum()]})
itm_res["Average Price"] = itm_res["Average Price"].map("${:.2f}".format)
itm_res["Total Revenue"] = itm_res["Total Revenue"].map("${:.2f}".format)
#Results
itm_res
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#New copied df that isolates values we want to analyze
#gndr - Gender
gndr_df = purchase_data[["SN","Gender"]].copy()

#Sort and remove duplicates
gndr_df.sort_values("SN",inplace = True)
gndr_df.drop_duplicates(keep = 'first', inplace = True)

#Results
res_gndr = pd.DataFrame(gndr_df["Gender"].value_counts())
res_gndr = res_gndr.rename(columns = {"Gender":"Total Count"})
res_gndr["Percentage of Players"] = round((res_gndr["Total Count"]/gndr_df["Gender"].count())*100,2)
res_gndr["Percentage of Players"] = res_gndr["Percentage of Players"].map("{:.2f}%".format)
res_gndr
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#gp - Gender Purchase
gp_df = purchase_data[["SN","Gender","Price"]]
gp_group = gp_df.groupby("Gender")

#Results
res_gp = pd.DataFrame(gp_group["Price"].count())
res_gp["Average Purchase Price"] = round(gp_group["Price"].mean(),2)
res_gp["Total Purchase Value"] = round(gp_group["Price"].sum(),2)
res_gp["Avg Total Purchase per Person"] = round(res_gp["Total Purchase Value"]/res_gndr["Total Count"],2)

#Formatting
res_gp["Average Purchase Price"] = res_gp["Average Purchase Price"].map("${:.2f}".format)
res_gp["Total Purchase Value"] = res_gp["Total Purchase Value"].map("${:.2f}".format)
res_gp["Avg Total Purchase per Person"] = res_gp["Avg Total Purchase per Person"].map("${:.2f}".format)
res_gp
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Define bins for later
bin_age = [0,9,14,19,24,29,34,39,100]
bin_label = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

#Extracting values for analysis
age_df = purchase_data[["SN","Age"]].copy()
age_df.sort_values("SN",inplace = True)
age_df.drop_duplicates(keep = 'first', inplace = True)

#Results
res_age = pd.cut(age_df["Age"], bin_age, labels=bin_label,include_lowest = True).value_counts().sort_index()
res_age = pd.DataFrame(res_age).rename(columns = {"Age":"Total Count"})
res_age["Percentage of Players"] = round((res_age["Total Count"]/age_df["Age"].count())*100,2) 
res_age["Percentage of Players"] = res_age["Percentage of Players"].map("{:.2f}%".format)
res_age
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#New extraction and cuts
agebuy_df = purchase_data[["SN","Age","Price"]].copy()
agebuy_df["Age Range"] = pd.cut(agebuy_df["Age"], bin_age, labels=bin_label,include_lowest = True)
age_group = agebuy_df.groupby(["Age Range"])

#Results
res_pa = pd.cut(agebuy_df["Age"], bin_age, labels=bin_label,include_lowest = True).value_counts().sort_index()
res_pa = pd.DataFrame(res_pa).rename(columns = {"Age":"Total Count"})
res_pa["Average Purchase"] = age_group["Price"].mean()
res_pa["Total Purchase"] = age_group["Price"].sum()
res_pa["Average Total Purchase"] = age_group["Price"].sum()/age_group["SN"].nunique()

#Formatting
res_pa["Average Purchase"] = res_pa["Average Purchase"].map("${:.2f}".format)
res_pa["Total Purchase"] = res_pa["Total Purchase"].map("${:.2f}".format)
res_pa["Average Total Purchase"] = res_pa["Average Total Purchase"].map("${:.2f}".format)
res_pa
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Extract SN data to distinguish who spends the most
sn_df = purchase_data[["SN","Price"]].copy()
sn_df.head()
sn_group = sn_df.groupby("SN")

#Results
res_sn = sn_group.count().rename(columns = {"Price":"Purchase Count"})
res_sn["Average Purchase Price"] = sn_group["Price"].mean()
res_sn["Total Purchase Value"] = sn_group["Price"].sum()
res_sn = res_sn.sort_values(["Total Purchase Value"], ascending=False)

#Formatting
res_sn["Average Purchase Price"] = res_sn["Average Purchase Price"].map("${:.2f}".format)
res_sn["Total Purchase Value"] = res_sn["Total Purchase Value"].map("${:.2f}".format)
res_sn.head() #Top 5 spenders
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Extract item information for analysis
item_df = purchase_data[["Item ID","Item Name","Price"]]
item_group = item_df.groupby(["Item ID","Item Name"])

#Results
res_item = item_group.count()
res_item["Count"] = item_group.count()
res_item["Price"] = item_group["Price"].mean()
res_item["Total Purchase Price"] = item_group["Price"].sum()
res_item = res_item.sort_values(["Count"], ascending=False)

#Formatting 
res_item["Price"] = res_item["Price"].map("${:.2f}".format)
res_item["Total Purchase Price"] = res_item["Total Purchase Price"].map("${:.2f}".format)
res_item.head()
```
